```python
## apply, map의 용도는 df의 컬럼의 값을 특정한 처리 후, df의 컬럼으로 다시 갱신하거나, 파생변수를 만들어
## 집어넣기 위한 것.
## ==> apply, map의 함수로 사용되는 것은 입력값이 시리즈의 값 하나를 처리하는 단위로 만들어야 함.
## ==> return을 반드시 넣어서 함수로 만들어야 함.
def str_count(x): ## x는 시리즈(리스트)의 하나의 값을 처리하는 단위(값하나)
    if len(x) >= 15:
        return "long"
    else:
        return "short"
```
