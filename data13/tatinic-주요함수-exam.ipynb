```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')
# warnings.filterwarnings(action='default')
%matplotlib inline
```

1. p-class columns update ==> 1:low, 2 ~ 3:middle , 4 ~ 5:high

```python
def pclass_convert(pclass):
    if pclass==1:
        return "low"
    elif 2<=pclass<=3:
        return "middle"
    else:
        return "high"
```

```python
lambda pclass : "low" if pclass==1 else("middle" if 2<=pclass<=3 else "high")
```

```python
# df['pclass']=df['pclass'].apply(pclass_convert)
df['pclass']=df['pclass'].apply(lambda pclass : "low" if pclass==1 else("middle" if 2<=pclass<=3 else "high"))
df
```

2. name 으로 파생변수 name_result를 만들어라. 글자수를 세어 15이상이면 long, 그외에는 short

```python
def name_length(name):
    if len(name)>=15:
        return "long"
    else:
        return "short"
```

```python
lambda name : "long" if len(name)>=15 else "short"
```

```python
# df['name_result']=df['name'].apply(name_length)
df['name_result']=df['name'].apply(lambda name : "long" if len(name)>=15 else "short")
```

3. survived 1 이면 생존, 0이면 사망, survived columns update

```python
def survived (survive):
    if survive==1:
        return "생존"
    else:
        return "사망"
```

```python
# df['survived']=df['survived'].apply(survived)
df['survived']=df['survived'].apply(lambda survive :"생존" if survive==1 else "사망")
```

age 10대, 20대, 30대, 40대, 60대 이상으로 columns update

```python
def ages (age):
    if age<10:
        return "유아"
    elif 10<=age<20:
        return "10대"
    elif 20<=age<30:
        return "20대"
    elif 30<=age<40:
        return "30대"
    elif 40<=age<50:
        return "40대"
    elif 50<=age<60:
        return "50대"
    
    else:
        return "60대 이상"
```

```python
lambda age: "유아" if age<10 else("10대" if 10<=age<20 else("20대" if 20<=age<30 else ("30대" if 30<=age<40 else("40대" if 40<=age<50 else("50대" if 50<=age<60 else "60대 이상")))))
```

```python
df['age']=df['age'].apply(ages)
# df['age']=df['age'].apply(ages)
```

name에 Mr, Dr을 포함하고있으면 남자, Miss를 포함하고있으면 여자, 나머지는 기타. 
파생변수 name_result2를 만들고 각각을 카운트. 남자 몇명, 여자 몇명?

```python
def name_gender(name):
    if "Mr" in name or "Dr" in name:
        return "남자"
    elif "Miss" in name:
        return "여자"
    else:
        return "기타"
    
```

```python
lambda name : "남자" if "Mr" in name or "Dr" in name else("여자" if "Miss" in name else "기타")
```

```python
df['name_result2']=df['name'].apply(lambda name : "남자" if "Mr" in name or "Dr" in name else("여자" if "Miss" in name else "기타"))
```
