```python
## 하드보팅/소프트보팅해보세요.
```

```python
train_df.drop(['district'], axis=1, inplace = True)
test_df.drop(['dong'], axis=1, inplace = True)
```

```python
dtree = DecisionTreeClassifier(
    criterion='entropy', 
    max_depth=25,
    random_state=35
)
```

```python
knn = KNeighborsClassifier(
    n_neighbors=10,
    n_jobs=None
)
```

```python
svm = SVC(
    C=0.1, 
    gamma=0.003,
    kernel='rbf',
    probability=True ##각 라벨링값으로 될 가능성 구해주세요.!
)
```

```python
## 하드 보팅
voting_clf1 = VotingClassifier(
    estimators= [
        ('decision tree', dtree), 
        ('knn', knn), 
        ('svm', svm)
    ],
    weights=[1, 1, 1],
    voting='hard'
)
```

```python
train_df.columns
```

```python
test_df.columns
```

```python
voting_clf1.fit(train_df[['longitude','latitude']], train_df['label'])
hard_voting_predicted = voting_clf1.predict(test_df[['longitude','latitude']])
accuracy_score(test_df['label'], hard_voting_predicted)
```

```python
## 소프트 보팅
voting_clf2 = VotingClassifier(
    estimators= [
        ('decision tree', dtree), 
        ('knn', knn), 
        ('svm', svm)
    ],
    weights=[1, 1, 1],
    voting='soft'
)
```

```python
voting_clf2.fit(train_df[['longitude','latitude']], train_df['label'])
soft_voting_predicted = voting_clf2.predict(test_df[['longitude','latitude']])
accuracy_score(test_df['label'], soft_voting_predicted)
```

```python
### 그래프로 그려보세요. accuracy(dtree, knn, svm, hard voting, soft voting)
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x = np.arange(2)
plt.bar(x, height= [
                    accuracy_score(test_df['label'], hard_voting_predicted),
                    accuracy_score(test_df['label'], soft_voting_predicted)
            ])
plt.xticks(x, ['hard voting','soft voting']);
```
