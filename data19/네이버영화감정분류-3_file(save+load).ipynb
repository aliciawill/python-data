```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from konlpy.tag import Okt
```

```python
okt = Okt()
```

```python
count_vec = CountVectorizer(tokenizer= okt.morphs)
```

```python
tfidf_2 = TfidfVectorizer(
                        ngram_range=(1, 2), 
                        min_df = 3, 
                        max_df = 0.9)
tfidf_2
# TfidfVectorizer(
#     *,
#     input='content',
#     encoding='utf-8',
#     decode_error='strict',
#     strip_accents=None,
#     lowercase=True,
#     preprocessor=None,
#     tokenizer=None,
#     analyzer='word',
#     stop_words=None,
#     token_pattern='(?u)\\b\\w\\w+\\b',
#     ngram_range=(1, 1),
#     max_df=1.0,
#     min_df=1,
#     max_features=None,
#     vocabulary=None,
#     binary=False,
#     dtype=<class 'numpy.float64'>,
#     norm='l2',
#     use_idf=True,
#     smooth_idf=True,
#     sublinear_tf=False,
# )
# analyzer : {'word', 'char', 'char_wb'} or callable, default='word'
#     Whether the feature should be made of word or character n-grams.
#     Option 'char_wb' creates character n-grams only from text inside
#     word boundaries; n-grams at the edges of words are padded with space.    
```

```python
tfidf = TfidfVectorizer(tokenizer= okt.morphs , 
                        ngram_range=(1, 2), 
                        min_df = 3, 
                        max_df = 0.9)
tfidf
# max_df : 정수 또는 [0.0, 1.0] 사이의 실수. 디폴트 1
#  최대 몇개 문서에 걸쳐 포함된 단어까지 단어 꾸러미에 담아 사용할 것인지
# min_df : 정수 또는 [0.0, 1.0] 사이의 실수. 디폴트 1
#  최소 몇개 문서에 걸쳐 포함된 단어를 단어 꾸러미에 담아 사용할 것인지
## ngram-range(1,2): 한글자가 따로 나오거나, 두 글자가 같이 나오는 빈도수의 matrix
```

```python
tfidf_2.fit(train_df['document'])
### 사전만들때나, 스케일링시 fit()시킬 때, 반드시  train데이터를 사용한다.!
### test데이터만을 가지고 fit()시키지 않는다.!
```

```python
train_df['document']
```

```python
train_tfidf_df = tfidf_2.transform(train_df['document'])
train_tfidf_df 
##df의 컬럼으로 들어갈 수 있도록 one-hot encoding하는 것.(DTM만드는 중)
```

```python
test_tfidf_df = tfidf_2.transform(test_df['document'])
test_tfidf_df
# sparse matrix(희소행렬, DTM)
```

```python
len(test_tfidf_df.toarray()) #2nd array
```

```python
from sklearn.linear_model import LogisticRegression

SA_lr = LogisticRegression(random_state = 0)
```

```python
train_tfidf_df
```

```python
train_df['label']
```

```python
SA_lr.fit(train_tfidf_df, train_df['label'])
```

```python
test_predict = SA_lr.predict(test_tfidf_df)
```

```python
from sklearn.metrics import accuracy_score

print('감성 분석 정확도 : ', round(accuracy_score(test_df['label'], test_predict), 3))
```
