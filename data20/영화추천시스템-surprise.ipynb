```python
## 고유값(특이값)분해 방법
from surprise import SVD
## SGD(Stochastic 통계적인 Gradient 경사 하강법)
from surprise import Dataset, Reader
from surprise import accuracy
from surprise.model_selection import train_test_split
```

```python
data = Dataset.load_builtin('ml-100k')
data #DatasetAutoFolds
```

```python
trainset, testset = train_test_split(data, test_size=0.25, random_state=0)
```

```python
trainset
```

```python
testset[1] #882번 사람이 291번 영화에 4.0 평점을 부여함.
```

```python
testset[:5]
```

```python
algo = SVD()
```

```python
algo.fit(trainset)
```

```python
prediction = algo.test(testset)[1]
prediction #하나의 영화에 대한 예측값은 Prediction객체!
```

```python
predictions = algo.test(testset)[:5]
predictions 
```

```python
prediction
```

```python
'userid: ' + prediction.uid + \
', iid: ' + prediction.iid + \
' ==> est: ' + str(prediction.est)
```

```python
predictions
```

```python
## predictions의 모든 예측값 출력
[one.uid + ', ' + one.iid + ', ' + str(one.est) for one in predictions]
```

```python
accuracy.rmse(predictions)
```

```python
predictions2 = algo.test(testset)
predictions2 
```

```python
## predictions2의 모든 예측값 출력
result2 = [one.uid + ', ' + one.iid + ', ' + str(one.est) for one in predictions2]
result2
```

```python
len(result2)
```

```python
accuracy.rmse(predictions2) #오차의 평균값.!
```
